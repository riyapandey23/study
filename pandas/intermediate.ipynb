1. GroupBy : 
Used to aggregate data by categories.

```python
df = pd.DataFrame({
    'Employee':['Aman','Riya','Karan','Zoya','Neha','Arjun'],
    'Department':['IT','HR','IT','Finance','HR','IT'],
    'Salary':[70000,50000,80000,90000,55000,72000]
})
```

```python
df.groupby('Department')['Salary'].mean().reset_index()
```

# Query (cleaner filtering)

```python
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 35, 40],
    'Department': ['IT', 'HR', 'IT', 'Finance'],
    'Salary': [50000, 60000, 70000, 80000]
}

df = pd.DataFrame(data)

# Filter employees with Age > 30
df.query('Age > 30')
```

```python
df.query('Salary > 60000')
df.query('Department == "IT" & Age < 35')
df.query('~( Department == "Finance" )')
df.query('Department == "HR" | Salary > 70000')
```

```python
# In query(), you can reference Python variables by adding @ before the variable name:
dept_var = "IT"         # Python variable dept_var.
salary_limit = 65000        # Python variable salary_limit.

df.query('Department == @dept_var & Salary > @salary_limit')
```

```python
dept_var = "HR"
df.query('Department == @dept_var')

age_limit = 3
df.query('Age > @age_limit')

dept_var = "IT"
salary_limit = 60000
df.query('Department == @dept_var & Salary > @salary_limit')

dept_var1 = "HR"
salary_limit = 80000
df.query('Department == @dept_var | Salary > @salary_limit')

dept_var = "Finance"
df.query('Department != @dept_var')

dept_var = "IT"
age_min = 30
df.query('(Department == @dept_var) & (Age >= @age_min) & (Age <= @age_max)')
```
