# Query (cleaner filtering)

````python
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 35, 40],
    'Department': ['IT', 'HR', 'IT', 'Finance'],
    'Salary': [50000, 60000, 70000, 80000]
}

df = pd.DataFrame(data)

# Filter employees with Age > 30
df.query('Age > 30')
````

````python
df.query('Salary > 60000')
df.query('Department == "IT" & Age < 35')
df.query('~( Department == "Finance" )')
df.query('Department == "HR" | Salary > 70000')
````

````python
# In query(), you can reference Python variables by adding @ before the variable name:
dept_var = "IT"         # Python variable dept_var.
salary_limit = 65000        # Python variable salary_limit.

df.query('Department == @dept_var & Salary > @salary_limit')
````

# Apply function

````python
import pandas as pd

df = pd.DataFrame({
    'Name': ['Alice','Bob','Charlie'],
    'Math': [80, 60, 90],
    'Science': [70, 65, 95]
})

df['Salary'] = [60000,50000,70000]

df
````

````python
df['Math'] = df['Math'].apply(lambda x : x+10 if x <=70 else x)
df['Total'] = df.apply(lambda row: row['Math'] + row['Science'], axis=1)
#df = df.rename(columns={'total':'Total'})
df['Result'] = df['Total'].apply(lambda x : 'Pass' if x >= 150 else 'Fail')
df['Tax'] = df['Salary'].apply(lambda x: x * 0.10)
````

````python
df['Total'] = df.apply(lambda x: x['math'] + x['science'] + x['english'] , axis=1)      #This is a apply function to work with multi columns.
df= df.rename(columns={'Total':'total'})
print(df)
````

````python
def category(percent):
    if percent >= 90:
        return "90%+"
    elif percent >= 80:
        return "80%-90%"
    elif percent >= 70:
        return "70%-80%"
    elif percent >= 60:
        return "60%-70%"
    else:
        return "Below 60"


df['num_percent'] = df['total'] / 300 * 100

df['percentage'] = df['num_percent'].apply(category)

df['result'] = df['num_percent'].apply(lambda x: 'Pass' if x >= 35 else 'Fail')

print(df)
````

# Transform

````python
import pandas as pd

df = pd.DataFrame({
    'Name': ['A', 'B', 'C', 'D', 'E'],
    'Dept': ['IT', 'IT', 'HR', 'HR', 'HR'],
    'Salary': [50, 60, 55, 55, 65]
})


print(df)
````

````python
df['Rank'] = df['Salary'].transform('rank' , method = 'dense' , ascending=False).astype('Int64') + 100          #ranking starts with 101
df = df.sort_values(by='Salary' , ascending=False)
print(df)
````

````python
df['Rank'] = df.groupby('Dept')['Salary'].transform('rank', ascending=False)
print(df)
````

````python
df['Rank'] = df['Salary'].rank(ascending=False)
print(df)
````

````python
df['Rank'] = df['Salary'].rank(ascending=False, method='dense')
print(df)
````

````python
df['Rank'] = df['Salary'].rank(ascending=False, method='dense')
df = df.sort_values(by='Salary', ascending=False)
print(df)
````

# Pivot Table
A Pivot Table is used to summarize and analyze data easily.

👉 You use it when you want to:

Summarize data |
Group data |
Aggregate values (sum, avg, count, etc.) |
Reshape data into a report format

👉 It helps you answer questions like:

Total salary per department?
Average marks per student?
Count of employees in each city?

## 📊 Pro Tips (Interviews)

### ✅ Use `pivot_table()` when:
- You need **Excel-like summaries**
- You want **multiple aggregations** (mean, sum, count, etc.)
- You need a **clean, structured output**

#### 🔹 Example:
```python
import pandas as pd

df = pd.DataFrame({
    'Dept': ['IT','HR','IT','HR','IT'],
    'Salary': [50000,60000,70000,80000,90000]
})

pivot = pd.pivot_table(
    df,
    values='Salary',
    index='Dept',
    aggfunc=['mean','max','count']
)

print(pivot)

````python
import pandas as pd

df = pd.DataFrame({
    'Name': ['A','B','C','D'],
    'Dept': ['IT','HR','IT','HR'],
    'Salary': [50000,60000,70000,80000],
    'Gender': ['Male','Female','Male','Female']
})

pivot = df.pivot_table(values='Salary', index='Dept', aggfunc='mean')
print(pivot)
````

values

👉 Which column to calculate

index

👉 Row grouping

columns

👉 Column grouping (like Excel pivot)

````python
pivot = df.pivot_table(values='Salary' , index ='Dept' , columns='Gender' , aggfunc=['sum','mean','max','min'])
pivot = pivot.fillna(0)
pivot = pivot.stack(1)  # move aggfunc to rows # by default agg is on columns
print(pivot)
````

❓ 3. “Why use pivot_table instead of groupby?”

This is the REAL question 👇

🧠 Truth:

👉 groupby() = engine (logic)

👉 pivot_table() = presentation (report)

````python
import pandas as pd 

df = pd.DataFrame({
    'Dep' : ['IT','IT','HR','CS'],
    'Salary' : [50000,550000,25000,60000]
})
print(df)
````

# .at → Fast scalar access (single value)
Key points:

Only for single cell |
Faster than .loc |
Cannot select multiple rows
# .iat → Fast position-based scalar
Key points:

Like .iloc but for single value only |
Fastest access method

````python
#df.loc[1, 'Salary'] = 60000   
df.at[1, 'Salary'] = 60000           #Fast scalar access (single value) than loc 
print(df)
````

````python
df.iat[1 ,0] = 'CS'
print(df)
````

index = group by row category

“Group salaries by department (Dep)” [index = GROUP BY column]

Columns = split into categories

This is used when you want to break a group into sub-groups horizontally

“Group by Dep, then split each department into Male/Female columns”

````python
pivot = df.pivot_table(
    values='Salary',
    index='Dep',
    aggfunc=['mean','max','min']
).reset_index()

pivot.columns = ['Dep', 'Mean', 'Max', 'Min']

print(pivot)
````
