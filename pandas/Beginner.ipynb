Grouping Conditions (Very Important)

() is also used for logical conditions in pandas.

Because Python needs to know which condition to evaluate first.

```python
data[(data['Age'] > 20) & (data['City'] == 'Chicago')]
data[(data['Age'] < 25 ) | (data['City'] == 'Chicago')]
data[data['City'] != 'Boston']
```

```python
data[data['Name'].str.startswith('R')]
data[data['Name'].str.endswith('a')]
data[data['Name'].str.contains('ee')]
```
