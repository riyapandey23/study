```python
import pandas as pd
```

# Two Main Data Structures
1. Series:
A single column of data

```python
data = [1,2,3,4,5]
df = pd.Series(data, name='numbers', index=['a','b','c','d','e'])
print(df,end='\n\n')

s = df.to_frame()               # the index is preserved automatically.
print(s,end='\n\n')

df = pd.DataFrame(data)         # the index is generated automatically.
print(df)
```

```python
df = pd.DataFrame(data, columns=['numbers'] , index = ['a','b','c','d','e'])        # the index is generated automatically. unless u specify it.
print(df)
```

2. DataFrame:
A table (rows and columns)

```python
data1 = pd.DataFrame({
    'name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank', 'Grace', 'Helen'],
    'age': [25, 30, 35, 28, 22, 40, 27, 33],
    'height_cm': [165, 175, 180, 170, 160, 178, 168, 172],
    'weight_kg': [55, 78, 85, 72, 50, 90, 60, 68],
    'marks_math': [88, 92, 79, 85, 90, 75, 95, 82],
    'marks_science': [91, 89, 84, 88, 93, 80, 97, 86],
    'city': ['New York', 'Los Angeles', 'Chicago', 'Boston', 'Seattle', 'Houston', 'Denver', 'Miami']
})

data1.to_csv('students.csv', index=False, header=True)
```

```python
df = data1.rename(columns={
    'name' : 'Name',
    'age'	: 'Age',
    'height_cm'	: 'Height',
    'weight_kg'	: 'Weight',
    'marks_math' :	'Maths',
    'marks_science' :  'Science',
    'city' : 'City'
})

df
```

```python
df.to_csv('students.csv', index = False , header=True)

print(df)   
```

📌drop() does not modify the original dataframe unless you:

Option 1 — Assign it back

Option 2 — Use inplace=True

```python
df.drop('Science' ,axis=1, inplace=True)   # axis = 1 means column and inplace = True means it will change the original dataframe.
print(df)
```

```python
df.drop(columns=['Maths'], inplace=True)   # another way to drop a column. Better cleaner syntax.
df
```

```python
df['Maths'] = [88, 92, 79, 85, 90, 75, 95, 82]  # adding a new column to the dataframe.
df['Science'] = [91, 89, 84, 88, 93, 80, 97, 86]  # adding a new column to the dataframe.
print(df)
```

```python
df.drop(index=[1,3], inplace = True)   # dropping rows by index.
print(df)
```

```python
df.index = range(1, len(df) +1)   # resetting the index after dropping rows.
print(df)
```

```python
col = df.pop('Science')
df.insert(4, 'Science', col)   # inserting the column back to its original position.
print(df)
```

```python
print(list(data1.columns))
print(list(df.columns))
```

```python
col = df.pop('City')
df['City'] = col # adding the column back to the end of the dataframe.
print(df)
```

Updating 

```python
df.loc[2,'Age'] = 40    #Assigns the value  (Updating single value)
df 
```

```python
df.loc[2,'Age'] == 30       #comparison operator 
df
```

```python
df.loc[df['Name'] == 'Charlie' , 'Maths'] = 85 
df.loc[df['Name'] == 'Helen' , 'Weight'] = 62        #Updating Multiple Values
df
```

```python
df['Age'] = df['Age'] + 1       #Updating an Entire Column
df
```

```python
df.loc[df['Name'] == 'Frank' , 'Weight'] = 80       #Update using Condition
df
```

```python
df['Result'] = ''
df.loc[(df['Science'] >= 90) & (df['Maths'] >= 90) , 'Result'] = 'A+'
df
```

```python
df['Total'] = df['Science'] + df['Maths']
df.loc[df['Total'] >= 180, 'Result'] = 'A+'
df.loc[(df['Total'] >= 140) & (df['Total'] < 180) , 'Result' ] = 'A'
```
