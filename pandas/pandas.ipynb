.strip() is a string method

Removes leading and trailing spaces from strings

```python
new_data = {
    'Name' : ['  Riya','Shreeram    ','Ram'],
    'Age' : [21,21,20]
}

new_data = pd.DataFrame(new_data)
print("Before strip:\n",new_data)


data_clean = new_data['Name'].str.strip()
print("\nAfter strip:\n",data_clean)
```
