```python
import pandas as pd

data = {
    'Employee': ['Aman','Riya','Karan','Zoya','Arjun','Meera','Kabir','Archita','Rahul','Pushpa'],
    'Department': ['IT','HR','IT','Finance','HR','IT','Finance','IT','HR','Finance'],
    'City': ['Mumbai','Delhi','Mumbai','Delhi','Bangalore','Mumbai','Delhi','Bangalore','Mumbai','Bangalore'],
    'Salary': [70000,50000,65000,90000,52000,72000,88000,72000,51000,91000],
    'Experience': [4,3,5,6,3,4,7,4,2,8],
    'Performance': [4,3,4,5,3,4,5,4,2,5]
}

df = pd.DataFrame(data)
print(df)
```

```python
df.drop('City',axis=1)
df.drop(['Salary','Performance'],axis=1)
df.drop(3,axis=0)
df.drop([2,5,7],axis=0)
df.drop(['Experience'],axis=1,inplace=True)
df.drop(df.iloc[0:3].index)
df.drop(df[df['Department'] == 'HR'].index,axis=0,inplace=True)
df.reset_index(drop=True,inplace=True)
df.drop(df.tail(3).index,axis=0,inplace=True)
df.drop(index=[1,4],columns='Salary')
df.drop(columns=['Department','City'])
print(df)
```
