✅ Deep Copy vs Shallow Copy

🔥 Why all look same except s?

Because your list is 1D (no nested objects).

👉 Important rule:  
For simple lists (no nested lists)  
👉 shallow copy ≈ deep copy                
           

```
💡 Important Rule   
🔹 copy() (Shallow Copy)    
Copies outer list only          

Inner objects are still shared  
🔹 deepcopy()   
Copies everything (nested too)  
```

````python
import copy as cp 
a = [[1,2,3],4,5,6,7]

b = cp.copy(a)
c = cp.deepcopy(a)

a.append(8)

print(a)
print(b)
print(c)
````

```
a  ──►  [  X  , 4 , 5 ]
          │
          ▼
       [1,2,3]
```
       
- `a` → outer list  
- `[1,2,3]` → inner list (separate object in memory)  
- `X` → reference (pointer) to the inner list
