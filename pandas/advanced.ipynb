## 🧠 Adding Values in Python Data Structures

| Operation | List | Tuple | Set | Dict |
|----------|------|------|-----|------|
| Add single | `append(x)` | ❌ Not allowed | `add(x)` | `d[key] = value` |
| Add multiple | `extend([a,b])` | ❌ Not allowed | `update([a,b])` | `update({...})` |
| Insert at index | `insert(i, x)` | ❌ Not allowed | ❌ Not supported | ❌ Not supported |
| Update existing | `l[i] = x` | ❌ Not allowed | ❌ Not index-based | `d[key] = new_value` |
| Create new (add) | ✅ Direct modify | `t = t + (x,)` | `s = s.union({...})` | `d = {**d, new_pair}` |

---

### 🚀 Quick Notes
- **List** → Most flexible (append, insert, update)  
- **Tuple** → Immutable (cannot add/update directly)  
- **Set** → Use `add()` / `update()` (no index)  
- **Dict** → Use `key = value` or `update()`  

## 🧠 Removing Values in Python Data Structures

| Operation | List | Tuple | Set | Dict |
|----------|------|------|-----|------|
| Remove specific value | `remove(x)` | ❌ Not allowed | `remove(x)` | `pop(key)` |
| Remove last element | `pop()` | ❌ Not allowed | ❌ Not supported | ❌ Not supported |
| Remove by index/key | `pop(i)` | ❌ Not allowed | ❌ Not supported | `pop(key)` |
| Delete using keyword | `del l[i]` | ❌ Not allowed | ❌ Not supported | `del d[key]` |
| Remove all | `clear()` | ❌ Not allowed | `clear()` | `clear()` |
| Safe remove | ❌ | ❌ | `discard(x)` | `pop(key, default)` |

---

### 🚀 Quick Notes
- **List** → Supports index & value removal  
- **Tuple** → Immutable (no removal)  
- **Set** → Use `remove()` or `discard()` (no index)  
- **Dict** → Remove using keys only  
- `discard()` → No error if value not found (very important ⚠️)  

✅ Deep Copy vs Shallow Copy

🔥 Why all look same except s?

Because your list is 1D (no nested objects).

👉 Important rule:  
For simple lists (no nested lists)  
👉 shallow copy ≈ deep copy                
           

```
💡 Important Rule   
🔹 copy() (Shallow Copy)    
Copies outer list only          

Inner objects are still shared  
🔹 deepcopy()   
Copies everything (nested too)  
```

```````python
import copy as cp 
a = [[1,2,3],4,5,6,7]           # Only nested list are allowed 

b = cp.copy(a)
c = cp.deepcopy(a)

a.append(8)

print(a)
print(b)                # Nothing changes
print(c)
```````

```
a  ──►  [  X  , 4 , 5 ]
          │
          ▼
       [1,2,3]

b  ──►  [  X  , 4 , 5 ]
          │
          ▼
       [1,2,3]

```
       
- `a` → outer list  
- `[1,2,3]` → inner list (separate object in memory)  
- `X` → reference (pointer) to the inner list  

👉 a and b have different outer lists  
👉 BUT both point to the same inner list

``````
🔥 That’s what this means:
👉 “Copies outer list only”

✔️ a and b are different lists

👉 “Inner objects are shared”

❗ Both use the SAME [1,2,3]
``````

```````python
import copy as cp 
a = [[1,2,3],4,5,6,7] 

b = cp.copy(a)
c = cp.deepcopy(a)

a[0][1] = 8

print(a)
print(b)
print(c)
```````
